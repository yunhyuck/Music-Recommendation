## TAG MERGE FUNCTION

```python
data = []

def find_multiple_value(arr, a): 
    global data
    length = len(arr) 
    for x in range(0,length-1):
        if List_06_26['USERID'][x] == a :
            for y in range(x+1,length): 
                if List_06_26['TIME'][x] == List_06_26['TIME'][y]: 
                    data.append(List_06_26['TIME'][x])
                    break 
    return list(set(data))
```
